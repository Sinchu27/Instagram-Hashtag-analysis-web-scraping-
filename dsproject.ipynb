```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

np.random.seed(0)
num_samples = 100
comments = np.random.randint(10, 1000, num_samples)
hashtags = np.random.randint(1, 20, num_samples)
popularity = (comments > 500).astype(int)  # 1 if comments > 500 else 0

# Create a DataFrame
df = pd.DataFrame({
    'Comments': comments,
    'Hashtags': hashtags,
    'Popularity': popularity
})

# Define features and target
X = df[['Comments', 'Hashtags']]
y = df['Popularity']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the SVM classifier
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = svm_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Print classification report
print('Classification Report:')
print(classification_report(y_test, y_pred))

# Print confusion matrix
print('Confusion Matrix:')
print(confusion_matrix(y_test, y_pred))

# Create a mesh grid for plotting decision boundary
h = .02  # step size in the mesh
x_min, x_max = X['Comments'].min() - 1, X['Comments'].max() + 1
y_min, y_max = X['Hashtags'].min() - 1, X['Hashtags'].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# Plot the decision boundary
plt.figure(figsize=(10, 6))
Z = svm_model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)

# Plot the training points
plt.scatter(X_train['Comments'], X_train['Hashtags'], c=y_train, edgecolors='k', cmap=plt.cm.coolwarm)
plt.xlabel('Comments')
plt.ylabel('Hashtags')
plt.title('SVM Decision Boundary')
plt.show()

# Function to check if a post is popular
def check_popularity(comments, hashtags, model):
    input_data = np.array([[comments, hashtags]])
    prediction = model.predict(input_data)
    return "Popular" if prediction[0] == 1 else "Not Popular"

# Take user input
user_comments = int(input("Enter the number of comments: "))
user_hashtags = int(input("Enter the number of hashtags: "))

# Check if the post is popular
result = check_popularity(user_comments, user_hashtags, svm_model)
print(f'The post with {user_comments} comments and {user_hashtags} hashtags is {result}.')
```

```python
import time
import random
import pandas as pd
import matplotlib.pyplot as plt
from selenium import webdriver
from bs4 import BeautifulSoup
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Function to scrape Instagram posts
def scrape_instagram(hashtag, num_posts=20):
    URL = f"https://www.instagram.com/explore/tags/{hashtag}/"

    # Initialize the WebDriver
    driver = webdriver.Chrome()

    # Open the URL
    driver.get(URL)

    # Allow some time for the page to load
    time.sleep(5)

    # Scroll down to load more posts
    for _ in range(num_posts):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5)

    # Parse the page source
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Close the WebDriver
    driver.quit()

    # Extract post links
    posts = soup.find_all('a', href=True)
    post_urls = []
    count = 0
    for post in posts:
        href = post['href']
        if '/p/' in href:
            post_urls.append(f"https://www.instagram.com{href}")
            count += 1
        if count == num_posts:
            break

    return post_urls


def generate_dummy_data(post_urls, hashtag):
    num_posts = len(post_urls)
    data = {
        'URL': post_urls,
        'Comments': [random.randint(10, 1000) for _ in range(num_posts)],
        'Hashtags': [random.randint(1, 20) for _ in range(num_posts)],
        'Popularity': [(random.randint(10, 1000) > 500) for _ in range(num_posts)],  # 1 if comments > 500 else 0
        'Hashtag': [hashtag] * num_posts
    }
    return pd.DataFrame(data)

# Function to train SVM model
def train_svm_model(df):
    # Define features and target
    X = df[['Comments', 'Hashtags']]
    y = df['Popularity']

    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Initialize and train the SVM classifier
    svm_model = SVC(kernel='linear', random_state=42)
    svm_model.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = svm_model.predict(X_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    print(f'Accuracy: {accuracy:.2f}')

    # Print classification report
    print('Classification Report:')
    print(classification_report(y_test, y_pred))

    # Print confusion matrix
    print('Confusion Matrix:')
    print(confusion_matrix(y_test, y_pred))
    
    return svm_model

# Function to check if a post is popular
def check_popularity(comments, hashtags, model):
    input_data = np.array([[comments, hashtags]])
    prediction = model.predict(input_data)
    return prediction[0]

# Main program
if __name__ == "__main__":
    # Ask user to enter five hashtags
    hashtags = [input(f"Enter hashtag {i+1} (without #): ") for i in range(5)]

    all_dfs = []
    for hashtag in hashtags:
        # Scrape Instagram posts for each hashtag
        post_urls = scrape_instagram(hashtag)

        
        df = generate_dummy_data(post_urls, hashtag)
        all_dfs.append(df)

    # Combine all DataFrames
    combined_df = pd.concat(all_dfs, ignore_index=True)

    # Train SVM model
    svm_model = train_svm_model(combined_df)

    # Calculate popularity for each hashtag
    popularity_dict = {}
    for hashtag in hashtags:
        df_hashtag = combined_df[combined_df['Hashtag'] == hashtag]
        popularity_score = df_hashtag.apply(lambda row: check_popularity(row['Comments'], row['Hashtags'], svm_model), axis=1).mean()
        popularity_dict[hashtag] = popularity_score

    # Determine the hashtag with the highest popularity
    most_popular_hashtag = max(popularity_dict, key=popularity_dict.get)
    print(f'The hashtag with the highest popularity is #{most_popular_hashtag} with a score of {popularity_dict[most_popular_hashtag]:.2f}.')

    # Visualization
    # Area Graph for Hashtags vs. Comments for combined data
    plt.figure(figsize=(10, 6))
    sorted_df = combined_df.sort_values(by='Hashtags')  # Sort the DataFrame by Hashtags
    plt.fill_between(sorted_df['Hashtags'], sorted_df['Comments'], color="skyblue", alpha=0.4)
    plt.plot(sorted_df['Hashtags'], sorted_df['Comments'], color="Slateblue", alpha=0.6, linewidth=2)
    plt.xlabel('Number of Hashtags')
    plt.ylabel('Comments')
    plt.title('Area Graph: Hashtags vs. Comments')
    plt.show()
```
