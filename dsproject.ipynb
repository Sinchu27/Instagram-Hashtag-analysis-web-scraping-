```python
import time
import random
import pandas as pd
import matplotlib.pyplot as plt
from selenium import webdriver
from bs4 import BeautifulSoup

# Ask user to enter the hashtag
HASHTAG = input("Enter the hashtag (without #): ")
URL = f"https://www.instagram.com/explore/tags/{HASHTAG}/"

# Initialize the WebDriver
driver = webdriver.Chrome()

# Open the URL
driver.get(URL)

# Allow some time for the page to load
time.sleep(5)

# Scroll down to load more posts (set range to 20)
for _ in range(20):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)

# Parse the page source
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Close the WebDriver
driver.quit()

# Extract post links
posts = soup.find_all('a', href=True)
post_urls = []
count = 0
for post in posts:
    href = post['href']
    if '/p/' in href:
        post_urls.append(f"https://www.instagram.com{href}")
        count += 1
    if count == 20:
        break

# Print the post URLs
for url in post_urls:
    print(url)

# Generate dummy data for visualization matching the length of post_urls
num_posts = len(post_urls)
data = {
    'URL': post_urls,
    'Likes': [random.randint(100, 10000) for _ in range(num_posts)],
    'Comments': [random.randint(10, 1000) for _ in range(num_posts)],
    'Hashtags': [random.randint(1, 20) for _ in range(num_posts)]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Visualization
plt.figure(figsize=(14, 7))

# Subplot for Likes
plt.subplot(1, 3, 1)
plt.barh(df['URL'], df['Likes'], color='skyblue')
plt.xlabel('Likes')
plt.title('Distribution of Likes')

# Subplot for Comments
plt.subplot(1, 3, 2)
plt.barh(df['URL'], df['Comments'], color='salmon')
plt.xlabel('Comments')
plt.title('Distribution of Comments')



plt.tight_layout()
plt.show()
```
